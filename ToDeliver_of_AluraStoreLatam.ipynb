# **VOY A CREAR EL TOP 4 PRODUCTOS POR TIENDA, PARA QUE SE GRAFIQUE MEJOR**

```python
top4_ventas_por_tienda = {}

for tienda, productos in dicc_ventasProductosTODAS.items():
    # Ordenamos los productos por ingreso de mayor a menor
    productos_ordenados = dict(sorted(productos.items(), key=lambda item: item[1], reverse=True))

    # Nos quedamos con los 4 primeros
    top4 = dict(list(productos_ordenados.items())[:4])

    # Guardamos en el nuevo diccionario
    top4_ventas_por_tienda[tienda] = top4
```

```python
for tienda, productos in top4_ventas_por_tienda.items():
    nombres = list(productos.keys())
    ingresos = list(productos.values())

    plt.figure(figsize=(8, 4))
    bars = plt.bar(nombres, ingresos, color='green')
    plt.title(f"Top 4 productos por ingreso en {tienda}")
    plt.xlabel("Producto")
    plt.ylabel("Ingreso")
    plt.xticks(rotation=45, ha='right', fontsize=10)

    # Mostrar los valores completos sobre cada barra
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, f"${yval:,.0f}", ha='center', va='bottom', fontsize=10)

    plt.tight_layout()
    plt.show()
```
